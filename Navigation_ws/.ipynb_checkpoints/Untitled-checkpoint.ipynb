```python
rows = df1.shape[0]
print("The rows value:",rows)
return1 = []
boughtat = 0
sellat = 0
count = 0
i = 0
j = 0
while (df1.at[j,'buy2'] != 1):
  return1.append(0)
  j = j + 1

while i < rows:
  if (df1.at[i,'buy2'] == 1):
    boughtat = df1.at[i,'close']
    return1.append(0)
    # when the final row has 1, the loop will terminate
    if (i == (df1.shape[0] -1) and df1.at[i, 'buy2']==1):
        break
    count = 1
    while (df1.at[i+count,'buy2'] == 0 ):
      sellat = df1.at[i+count,'high']
      variable = ((sellat - boughtat)/boughtat)*100
      # print(variable,"  ", boughtat,"  ",sellat)
      return1.append(variable)
      if ((i+count) == (df1.shape[0] - 1) and df1.at[(df1.shape[0]-1), 'buy2'] == 0):
        break
        # continue
      elif ((i+count) == (df1.shape[0] -1) and df1.at[(df1.shape[0]-1), 'buy2']==1):
        break
        # continue
      count = count + 1
    i = i + count
  else:
    i = i + 1


print("The lenght of list and row", len(return1),', rows:',rows)
df1['return1'] = return1
```
